```python
import pandas as pd
inmuebles = pd.read_csv('/content/drive/MyDrive/inmuebles_bogota.csv')
columnas = {'Baños': 'Banos', 'Área': 'Area'}
inmuebles = inmuebles.rename(columns=columnas)
```

```python
#Desafío
#Promedio de áreas de todos los inmuebles de los barrios en el dataset. El top 10.
```

```python
import pandas as pd
promedio = []
indices = []
barrios = inmuebles.Barrio.value_counts()
for n in barrios.index:
    indices.append(n)
    barrio = inmuebles[(inmuebles.Barrio == n)]
    promedio.append(barrio.Área.mean())
grafica = pd.Series(promedio,index=indices)
print(grafica.sort_values(ascending = False).head(10))
top10barrios = grafica.sort_values(ascending = False).head(10)
```

```python
top10barrios.sort_values(ascending = True).head(10).plot.barh()
```

```python
#Conteo, Mediana, Valores Minimo y Maximo de Habitaciones
```

**Conteo de Habitaciones y su Grafica**

```python
import pandas as pd
x = 0
conteo_habitaciones = inmuebles.Habitaciones.value_counts()
df = pd.DataFrame(columns=['Habitaciones','Inmuebles'],
                  index=range(len(conteo_habitaciones)))
for tupla in zip(conteo_habitaciones.index, conteo_habitaciones.values): #obtenemos la tupla en cada iteración
  df.iloc[x] = (tupla[0], tupla[1])
  x=x+1
tabla_habitaciones = df.style.hide(axis=0)
tabla_habitaciones
```

En la siguiente grafica se visualiza cuantos inmuebles con el mismo numero de habitaciones

```python
import matplotlib.pyplot as plt
df[['Habitaciones', 'Inmuebles']] = df[['Habitaciones', 'Inmuebles']].astype(int)
df.plot(x="Habitaciones", y=["Inmuebles"], kind="bar",figsize=(9,8))
plt.show()
```

Mediana de Habitaciones

```python
inmuebles.Habitaciones.sort_values(ascending = False).median()
```

```python
inmuebles.Habitaciones.median()
```

Meximos y Minimos de Habitaciones

```python
inmuebles_ordenados = inmuebles.Habitaciones.sort_values(ascending = True).values
inmuebles_ordenados
print(f"Maximo de habitaciones {inmuebles_ordenados[0]} y minimo de habitaciones {inmuebles_ordenados[len(inmuebles_ordenados)-1]}")
```

**Conteo de Valor**

**Mediana del valor de los inmuebles**

```python
inmuebles.Valor.values
```

```python
valores_inmuebles = []
for string_valor in inmuebles.Valor.values:
  new_string = ''.join(char for char in string_valor if char.isalnum())
  int_numero = int(new_string)
  valores_inmuebles.append(int_numero)
valores_inmuebles.sort(reverse=True)
datos_totales = len(valores_inmuebles)
datos_totales = datos_totales/2
datos_totales = int(datos_totales)
suma_datos_totales = valores_inmuebles[datos_totales]
suma_datos_totales
```
