```python
import pandas as pd
inmuebles = pd.read_csv('/content/drive/MyDrive/inmuebles_bogota.csv')
columnas = {'Baños': 'Banos', 'Área': 'Area'}
inmuebles = inmuebles.rename(columns=columnas)
```

```python
valor_separado = inmuebles.Valor.str.split(expand=True)
type(valor_separado)
inmuebles['Moneda'] = valor_separado[0]
inmuebles['Precio'] = valor_separado[1]
inmuebles['Precio'] = inmuebles['Precio'].str.replace('.','',regex=True)
inmuebles['Precio_Millon'] = inmuebles.Precio.astype('float')/1000000
```

```python
import pandas as pd
x = 0
longitud = len(lista_inmuebles)
df = pd.DataFrame(columns=['Barrio','Precio_Metro_Cuadrado'],
                  index=range(longitud))
for n in lista_inmuebles: 
  df.iloc[x] = (n, inmuebles[(inmuebles.Barrio == n)].Precio_Metro_Cuadrado.mean())
  x=x+1
df.Precio_Metro_Cuadrado.sort_values(ascending = False)
```

```python
df.head(30).Precio_Metro_Cuadrado.sort_values(ascending = False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
grafica = sns.stripplot(data=df.head(30), y="Barrio",x='Precio_Metro_Cuadrado')
plt.xlim((1,10))
grafica.set(title = "Precio de metro cuadrado de inmuebles filtradopor barrio en Bogota")
plt.savefig('/content/drive/MyDrive/valor_metros_cuadrados.png',format='png')
plt.show()
```
